```python
import numpy as np

data = np.genfromtxt("placement-data.csv", delimiter=",", skip_header=1)

X = data[:, :2]   
y = data[:, 2].reshape(-1, 1)  
```

```python
#Data nomralization is being done here
X = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
```

```python
#Backpropogation implementation
# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
# Initialize weights
np.random.seed(42)

input_neurons = 2
hidden_neurons = 3
output_neurons = 1

W1 = np.random.rand(input_neurons, hidden_neurons)
b1 = np.random.rand(1, hidden_neurons)

W2 = np.random.rand(hidden_neurons, output_neurons)
b2 = np.random.rand(1, output_neurons)
```

```python

learning_rate = 0.1
epochs = 5000
```

```python
# Training
for epoch in range(epochs):

    # -------- Forward Propagation --------
    hidden_input = np.dot(X, W1) + b1
    hidden_output = sigmoid(hidden_input)

    output_input = np.dot(hidden_output, W2) + b2
    y_pred = sigmoid(output_input)

    # -------- Error --------
    error = y - y_pred

    # -------- Backward Propagation --------
    d_output = error * sigmoid_derivative(y_pred)

    error_hidden = d_output.dot(W2.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # -------- Update Weights --------
    W2 += hidden_output.T.dot(d_output) * learning_rate
    b2 += np.sum(d_output, axis=0, keepdims=True) * learning_rate

    W1 += X.T.dot(d_hidden) * learning_rate
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate
```

```python
# Final prediction
predictions = (y_pred >= 0.5).astype(int)

print("First 10 Predictions:")
print(predictions[:10])
```

```python
accuracy = np.mean(predictions == y) * 100
print("Model Accuracy:", accuracy, "%")
```
